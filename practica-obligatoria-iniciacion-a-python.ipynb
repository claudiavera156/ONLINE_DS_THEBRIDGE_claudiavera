# EJERCICIO 1

```python
PRECIO_FAMILIAR = 8.95
PRECIO_MEDIANA = 9.90
```

```python
print("Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana ¿Cuántas pizzas familiares desea?")
```

```python
pizz_familiar = int(input())
```

```python
print("Estupendo, ¿cuántas pizzas medianas desea?")
```

```python
pizz_mediana = int(input())
```

```python
print(f"Estupendo, se están preparando {pizz_familiar} pizzas familiares y {pizz_mediana} pizzas medianas. Dígame su dirección")
```

```python
direcc = input()
```

```python
total_pizzas = pizz_familiar + pizz_mediana
monto_total = (pizz_familiar * PRECIO_FAMILIAR) + (pizz_mediana * PRECIO_MEDIANA)
```

```python
print(f"Le mandaremos las {total_pizzas} pizzas a la dirección {direcc}. Serán {monto_total:.2f}€. Muchas gracias por su pedido.")
```

# EJERCICIO 2

```python
entrada = input("Introduce 3 números enteros separados por comas (ejemplo: 1,2,3): ")
```

```python
numeros = [int(n) for n in entrada.split(",")]
```

```python
maximo = max(numeros)
minimo = min(numeros)
suma = sum(numeros)
```

```python
afirmacion = (maximo - minimo) == (suma / 5)
```

```python
print(f"Lista de números: {numeros}")
print(f"Número máximo: {maximo}")
print(f"Número mínimo: {minimo}")
print(f"Suma total: {suma}")
print(f"¿El máximo menos el mínimo es igual a la suma dividida por 5? {afirmacion}")
```
